## Diabetes Predictions

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor
```

```python
from sklearn.datasets import load_diabetes
```

```python
dataset= load_diabetes()
```

```python
dataset
```

```python
print(dataset['DESCR'])
```

```python
dataset.feature_names
```

```python
df_diabetes = pd.DataFrame(dataset.data, columns= dataset.feature_names)
df_diabetes.head()
```

```python
#independent features
X= df_diabetes

#target feature
y =dataset.target 
```

```python
y
```

```python
X.head()
```

```python
#split the train test data
X_train,X_test,y_train,y_test = train_test_split(X,y, test_size=0.30,random_state=10)
```

```python
X_train.head()
```

```python
#check the correlation metrics
X_train.corr()
```

```python
#its draw the heat map
plt.figure(figsize=(8,7))
sns.heatmap(X_train.corr(numeric_only= True), annot= True, cmap= 'crest')
plt.show()
```

# Decision Tree

```python
from sklearn.model_selection import GridSearchCV
```

```python
from sklearn.metrics import r2_score,mean_squared_error,mean_absolute_error
```

```python
#instantiate the model
regressor = DecisionTreeRegressor()

#intiate hyperparameters
params={'criterion':['squared_error', 'friedman_mse', 'absolute_error'],
       'splitter':['best', 'random'],
       'max_depth':[1,2,3,4,5,10,15,20,25],
       'max_features':['auto','sqrt', 'log2']
       }

#instantiate grid search cv
grid = GridSearchCV(regressor,
                   param_grid=params,
                   scoring='neg_mean_squared_error',
                   cv=5,
                   refit='accuracy')
```

```python
%%time
import warnings
warnings.filterwarnings('ignore')

#fit the model
grid.fit(X_train,y_train)
```

```python
grid.best_params_
```

```python
grid.best_score_
```

```python
y_pred = grid.predict(X_test)
```

```python
print(r2_score(y_test,y_pred))
print(mean_absolute_error(y_test,y_pred))
print(mean_squared_error(y_test,y_pred))
```

```python
#now plot your tree with best estimators
selectedmodel= DecisionTreeRegressor(criterion= 'squared_error',max_depth= 3,max_features= 'log2',splitter= 'best')
```

```python
selectedmodel.fit(X_train,y_train)
```

```python
#visualize your tree
from sklearn import tree
```

```python
plt.figure(figsize=(15,10))
tree.plot_tree(selectedmodel,filled = True)
```
